### Day 1 - Part One

We have been given a list of numbers, the challengeis to find 2 numbers that add up to 2020, multiply them and submit the product of those nuymbers as the answer. 

In order to get the numbers into Python I added them to a csv called 'day_1_numbers.csv'

````python
import pandas as pd
from itertools import combinations 

# upload csv of numbers into a dataframe
upload_path = r'C:\Users\andrew.morris\Documents\GitHub\murry_code\Advent of Code 2020\inputs\\' 
day_1_filename = "day_1_numbers.csv"
day_1_numbers_df = pd.read_csv(upload_path+day_1_filename) 

# convert dataframe column to a list
num_list = day_1_numbers_df['Number'].to_list()

print('--- creating all combinations of numbers ---')
num_comb = combinations(num_list, 2) # Get all combinations of 2 numbers

print('--- adding all combinations to a dataframe')
num_combo_df = pd.DataFrame(columns = ['num_combinations']) # create empty combo dataframe

# Append the obtained combinations to  all_player_combo_df
for i in list(num_comb): 
    num_dict = {'num_combinations': i}
    num_combo_df = num_combo_df.append(num_dict, ignore_index=True)

num_combo_df[['num1','num2']] = pd.DataFrame(num_combo_df.num_combinations.tolist(), index= num_combo_df.index)
num_combo_df['addition'] = num_combo_df.num1 + num_combo_df.num2 
num_combo_df['multiplication'] = num_combo_df.num1 * num_combo_df.num2 

answer = num_combo_df[(num_combo_df.addition == 2020)]
answer
````

### Day Two - Part One

We are given a list of containing a range a letter and a password. We have to figure out how many passwords are valid by checking to see if the letter occurs at the frequency within the range. 

As with the above the list has been saved as a csv with some of the columns split out to make manipulation easier. 

The answer we are looking for is the amount of valid passwords within the list

````python
import pandas as pd

# specifying upload path and filename
upload_path = r'C:\Users\andrew.morris\Documents\GitHub\murry_code\Advent of Code 2020\inputs\\' # HP
day_2_filename = "day_two_passwords.csv"

# create dataframe from file 
passwords_df = pd.read_csv(upload_path+day_2_filename) 

# create a slimline dataframe with only useful fields
slim_passwords_df = passwords_df[['min_num',
                                  'max_num',
                                  'letter',
                                  'password']].copy()

# convert to a list of lists
passwords_list = slim_passwords_df.values.tolist()
passwords_list

# creating empty lists for valid and invalid passwords
valid_list = []
invalid_list = []

for i in passwords_list:
    # count the frequency of the letter in the password, 
    # if greater than or equal to min and less than or equal to max then add to valid list else add to invalid list
    if i[3].count(i[2]) >= i[0] and i[3].count(i[2]) <= i[1]:
        valid_list.append(i)
    else:
        invalid_list.append(i)
           
print('Valid Passwords: '  + str(len(valid_list)))
print('Invalid Passwords: '+ str(len(invalid_list)))
````

# Day Two - Part Two

The criteria for a valid password has changed. The range is actually two positions within the password string, and the letter can only appear once in those positions.

so this which would have been valid:
1-4 l: alwl
where 'l' appears twice in 'lawl'
is now invalid as 'l' is at position 1 and position 4, where as it should only appear once out of those 2 positions

Challenge part two is to find out how many passwords are valid usaing this new logic


````python
import pandas

# specifying upload path and filename
upload_path = r'C:\Users\andrew.morris\Documents\GitHub\murry_code\Advent of Code 2020\inputs\\' # HP
day_2_filename = "day_two_passwords.csv"

# create dataframe from file 
passwords_df = pd.read_csv(upload_path+day_2_filename) 

# create a slimline dataframe with only useful fields
slim_passwords_df = passwords_df[['min_num',
                                  'max_num',
                                  'letter',
                                  'password']].copy()

# convert to a list of lists
passwords_list = slim_passwords_df.values.tolist()

# creating empty lists for valid and invalid passwords
valid_list2 = []
invalid_list2 = []

# Due to the way that the password csv has been formatted there is a space at the beginning of each string, 
# this means that the min and max values are also the same as the positions within the string 
# and no adjustment needs to be made for zero indexing  

for i in passwords_list:
    min_freq = i[0]
    max_freq = i[1]
    letter   = i[2]
    password = i[3]
    new_password_string = password[min_freq] + password[max_freq]
    new_pass_letter_count = new_password_string.count(letter)
    if new_pass_letter_count == 1:
        valid_list2.append(i)
    else:
        invalid_list2.append(i)
        
print('Valid Passwords: '  + str(len(valid_list2)))
print('Invalid Passwords: '+ str(len(invalid_list2)))        
````

### Day Three - Part One

Day three's challenge uses a path that looks like this:

```
...#.....#.......##......#.....
...#..................#........
....##....#.......#............
.........#.......#.......#.....
..#..............#.........#..#
.....#.........#....#....#....#
....##..........#.#.##.........
```
The dots represent clear path and the hashes represent trees and the same pattern repeats to the right many many times

The challenge is to calculate how many 'trees' you will hit if starting at the top left you go down a specific slope. 

The slope in question is the direct route taken if you go 3 across (to the right) and down one until you get to the bottom row. 



````python
import pandas as pd

pd.set_option('max_colwidth', 50)

# specifying upload path and filename
upload_path = r'C:\Users\andrew.morris\Documents\GitHub\murry_code\Advent of Code 2020\inputs\\' # HP
day_3_filename = "day_three_trees.csv"

# create dataframe from file 
trees_df = pd.read_csv(upload_path+day_3_filename) 
trees_df['row_number'] = trees_df.index
# copying the string 50 times to ensure the string is long enough to get to the bottom 
trees_df['extendedpath'] = trees_df.tree_path*50

# creating a list of all the paths
path_list = trees_df.values.tolist()

# creating blank string for the full path, 
# this will contain all the characters from going one down 3 across for each row until we get to the bottom
full_path_string = ''

for i in path_list:
    path_row = i[2]
    row_number = i[1]
    # identifying the position within the string we want to return  
    position = row_number * 3 + 1
    # converting posiotion to the zero based index format
    position_index = position - 1
    # identifying the character from the path string which is at that position
    element_at_position = path_row[position_index]
    # adding that character to the full path string
    full_path_string = full_path_string + element_at_position
#     print('path row:            ' + path_row)
#     print('row number:          ' + str(row_number))
#     print('position:            ' + str(position))
#     print('position index:      ' + str(position_index))
#     print('element at position: ' + element_at_position)
#     print('full_path_string:    '+ full_path_string)
    
full_path_string
print('Volume of trees in path: ' + str(full_path_string.count('#')))
````

### Day Three - Part Two

The second part of this challenge asks you to calculate the volume of trees you will hit if you go down each of the following slopes:
```
slope 1: Right 1, down 1.
slope 2: Right 3, down 1. (This is the slope you already checked.)
slope 3: Right 5, down 1.
slope 4: Right 7, down 1.
slope 5: Right 1, down 2.
```
Then the answer is the product of all of these numbers. 

````python
import pandas as pd

pd.set_option('max_colwidth', 50)

# specifying upload path and filename
upload_path = r'C:\Users\andrew.morris\Documents\GitHub\murry_code\Advent of Code 2020\inputs\\' # HP
day_3_filename = "day_three_trees.csv"

# example filename
# day_3_filename = "day_three_trees_example_data.csv"
# example answer In the above example
# these slopes would find 2, 7, 3, 4, and 2 tree(s) respectively;
# multiplied together, these produce the answer 336.

# create dataframe from file 
print('creating tree_df')
trees_df = pd.read_csv(upload_path+day_3_filename) 
trees_df['row_number'] = trees_df.index
# copying the string 50 times to ensure the string is long enough to get to the bottom 
trees_df['extendedpath'] = trees_df.tree_path*100
# creating a flag to see if the row should be included in slope 5 as it misses every other row (every odd numbered row)
trees_df['slope_5_row' ] = trees_df['row_number'].apply(lambda a: a % 2 == 0)

print('creating slope_5_trees_df')
slope_5_trees_df = trees_df[trees_df.slope_5_row == True].reset_index(drop = True)
slope_5_trees_df['row_number'] = slope_5_trees_df.index

# creating a list of all the paths
print('creating path_list')
path_list = trees_df.values.tolist()
print('creating slope_5_path_list')
slope_5_path_list = slope_5_trees_df.values.tolist()

print('-------------------------------')    

print('creating slope_1_full_path_string')
slope_1_full_path_string = ''
for i in path_list:
    path_row                 = i[2]
    row_number               = i[1]
    position                 = row_number * 1 + 1
    position_index           = position - 1
    element_at_position      = path_row[position_index]
    slope_1_full_path_string = slope_1_full_path_string + element_at_position
    
print('creating slope_2_full_path_string')    
slope_2_full_path_string = ''
for i in path_list:
    path_row                 = i[2]
    row_number               = i[1]
    position                 = row_number * 3 + 1
    position_index           = position - 1
    element_at_position      = path_row[position_index]
    slope_2_full_path_string = slope_2_full_path_string + element_at_position    

print('creating slope_3_full_path_string')
slope_3_full_path_string = ''
for i in path_list:
    path_row                 = i[2]
    row_number               = i[1]
    position                 = row_number * 5 + 1
    position_index           = position - 1
    element_at_position      = path_row[position_index]
    slope_3_full_path_string = slope_3_full_path_string + element_at_position   

print('creating slope_4_full_path_string')    
slope_4_full_path_string = ''
for i in path_list:
    path_row                 = i[2]
    row_number               = i[1]
    position                 = row_number * 7 + 1
    position_index           = position - 1
    element_at_position      = path_row[position_index]
    slope_4_full_path_string = slope_4_full_path_string + element_at_position   

print('creating slope_5_full_path_string')    
slope_5_full_path_string = ''
for i in slope_5_path_list:
    path_row                 = i[2]
    row_number               = i[1]
    position                 = row_number * 1 + 1
    position_index           = position - 1
    element_at_position      = path_row[position_index]
    slope_5_full_path_string = slope_5_full_path_string + element_at_position

print('-------------------------------')    
    
s1_trees = slope_1_full_path_string.count('#')
s2_trees = slope_2_full_path_string.count('#')
s3_trees = slope_3_full_path_string.count('#')
s4_trees = slope_4_full_path_string.count('#')
s5_trees = slope_5_full_path_string.count('#')
     
product_of_all_trees = s1_trees * s2_trees * s3_trees * s4_trees * s5_trees
    
print('Volume of trees in slope 1:  ' + str(s1_trees))
print('Volume of trees in slope 2:  ' + str(s2_trees))
print('Volume of trees in slope 3:  ' + str(s3_trees))
print('Volume of trees in slope 4:  ' + str(s4_trees))
print('Volume of trees in slope 5:  ' + str(s5_trees))
print('-------------------------------')    
print('All tree volumes multiplied: ' + str(product_of_all_trees))
````

### Day Five - Part One

We are given a list of boarding pass strings, these highlight the seating position through a series of 'F' / 'B' characters of 'L' / 'R'

The first 7 characters will either be F or B; these specify exactly one of the 128 rows on the plane (numbered 0 through 127). Each letter tells you which half of a region the given seat is in. Start with the whole list of rows; the first letter indicates whether the seat is in the front (0 through 63) or the back (64 through 127). The next letter indicates which half of that region the seat is in, and so on until you're left with exactly one row.

For example, consider just the first seven characters of FBFBBFFRLR:
```
Start by considering the whole range, rows 0 through 127.
F means to take the lower half, keeping rows 0 through 63.
B means to take the upper half, keeping rows 32 through 63.
F means to take the lower half, keeping rows 32 through 47.
B means to take the upper half, keeping rows 40 through 47.
B keeps rows 44 through 47.
F keeps rows 44 through 45.
The final F keeps the lower of the two, row 44.
```
The last three characters will be either L or R; these specify exactly one of the 8 columns of seats on the plane (numbered 0 through 7). The same process as above proceeds again, this time with only three steps. L means to keep the lower half, while R means to keep the upper half.

For example, consider just the last 3 characters of FBFBBFFRLR:
```
Start by considering the whole range, columns 0 through 7.
R means to take the upper half, keeping columns 4 through 7.
L means to take the lower half, keeping columns 4 through 5.
The final R keeps the upper of the two, column 5.
```

So, decoding FBFBBFFRLR reveals that it is the seat at row 44, column 5.

Every seat also has a unique seat ID: multiply the row by 8, then add the column. In this example, the seat has ID 44 * 8 + 5 = 357.

Here are some other boarding passes:
```
BFFFBBFRRR: row 70, column 7, seat ID 567.
FFFBBBFRRR: row 14, column 7, seat ID 119.
BBFFBBFRLL: row 102, column 4, seat ID 820.
```
