```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

filepath = "C:\\Users\\andrew.morris\\Documents\\GitHub\\murry_code\\Codecademy Lesson Notes\\test data\\"

#################################################

df = pd.read_csv(filepath+'07c_WorldCupMatches.csv') 

df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']

sns.set_style("whitegrid")
sns.set_context("poster", font_scale=.75)

f, axs = plt.subplots(figsize=(12,7))
ax = sns.barplot(data = df, 
                 x = 'Year', 
                 y = 'Total Goals',
                 estimator = sum)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
ax.set_title('Total World Cup Goals By Year')    

#################################################

df_goals = pd.read_csv(filepath+'07c_Goals.csv')

sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1.5)

f, ax2 = plt.subplots(figsize=(12,7))
ax2 = sns.boxplot(data = df_goals, x = 'year', y = 'goals',palette = 'Spectral')

ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
ax2.set_title('Box Plot To Show The Distribution of World Cup Goals By Year') 

#################################################

plt.show()
```
