### Basic Line Plots

You can create a basic line plot in plython by importing pyplot from matplotlib (the commonly use alias is ply)

##### syntax: 
```python
x_values = [0, 1, 2, 3, 4]
y_values = [0, 1, 4, 9, 16]

plt.show() ```

1. We are going to make a simple graph representing someoneâ€™s spending on lunch over the past week. First, define two lists, 'days' and 'money_spent', that contain the following integers:
    * days - 0,1,2,3,4,5,6  
    * money_spent - 10,12,12,10,14,22,24  
    

2. Plot days on the x-axis and money_spent on the y-axis using plt.plot().  


3. Show the plot using plt.show().

````python

from matplotlib import pyplot as plt

days = [0,1,2,3,4,5,6]
money_spent = [10,12,12,10,14,22,24]

plt.plot(days, money_spent)
plt.show()
````

### Linestyles

We can specify a different color for a line by using the keyword color with either an HTML color name or a HEX code:
```python
plt.plot(days, money_spent, color='green')
plt.plot(days, money_spent_2, color='#AAAAAA')
```

We can also make a line dotted or dashed using the keyword linestyle.
* Dashed:
```python
plt.plot(x_values, y_values, linestyle='--')```
* Dotted:
```python
plt.plot(x_values, y_values, linestyle=':')```
* No line:
```python
plt.plot(x_values, y_values, linestyle='')```

We can also add a marker using the keyword marker:
* A circle:
```python
plt.plot(x_values, y_values, marker='o')```
* A square:
```python
plt.plot(x_values, y_values, marker='s')```
* A star:
```python
plt.plot(x_values, y_values, marker='*')```

````python

from matplotlib import pyplot as plt

time = [0, 1, 2, 3, 4]
revenue = [200, 400, 650, 800, 850]
costs = [150, 500, 550, 550, 560]

plt.plot(time, revenue, color = 'purple', linestyle = '--')
plt.plot(time, costs, marker = 's', color = '#82edc9')
plt.show()
````

### Subplots

The command plt.subplot() needs three arguments to be passed into it:
* The number of rows of subplots
* The number of columns of subplots
* The index of the subplot we want to create

````python

from matplotlib import pyplot as plt

months = range(12)
temperature = [36, 36, 39, 52, 61, 72, 77, 75, 68, 57, 48, 48]
flights_to_hawaii = [1200, 1300, 1100, 1450, 850, 750, 400, 450, 400, 860, 990, 1000]


plt.subplot(1,2,1)
plt.plot(months, temperature)

# making the second subplot in the figure a scatterplot
plt.subplot(1,2,2)
plt.plot(months, flights_to_hawaii, linestyle = '', marker = 'o')

plt.show()
````

### Subplots Part II

You can adjust the spacing between and around subplots by using plt.subplots_adjust()
* left â€” the left-side margin, with a default of 0.125. You can increase this number to make room for a y-axis label
* right â€” the right-side margin, with a default of 0.9. You can increase this to make more room for the figure, or decrease it to make room for a legend
* bottom â€” the bottom margin, with a default of 0.1. You can increase this to make room for tick mark labels or an x-axis label
* top â€” the top margin, with a default of 0.9
* wspace â€” the horizontal space between adjacent subplots, with a default of 0.2
* hspace â€” the vertical space between adjacent subplots, with a default of 0.2

````python

from matplotlib import pyplot as plt

x = range(7)
straight_line = [0, 1, 2, 3, 4, 5, 6]
parabola = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]

plt.subplot(2,1,1)
plt.plot(x,straight_line)

plt.subplot(2,2,3)
plt.plot(x, parabola)

plt.subplot(2,2,4)
plt.plot(x, cubic)

plt.subplots_adjust(wspace = 0.35, bottom = 0.2)

plt.show()
````

#### Modify Ticks

Because our plots can have multiple subplots, we have to specify which one we want to modify. In order to do that, we call plt.subplot() in a different way.

you can rotate the ticklabels with 

```python
plt.xticks(rotation=90)
```

```python
ax = plt.subplot(1, 1, 1)
```
ax is an axes object, and it lets us modify the axes belonging to a specific subplot. Even if we only have one subplot, when we want to modify the ticks, we will need to start by calling either ax = plt.subplot(1, 1, 1) or ax = plt.subplot() in order to get our axes object.

```python
ax = plt.subplot()
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
ax.set_xticks([1, 2, 4])

ax = plt.subplot()
plt.plot([1, 3, 3.5], [0.1, 0.6, 0.8], 'o')
ax.set_yticks([0.1, 0.6, 0.8])
ax.set_yticklabels(['10%', '60%', '80%'])
```

````python
from matplotlib import pyplot as plt

month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep","Oct", "Nov", "Dec"]

months = range(12)
conversion = [0.05, 0.08, 0.18, 0.28, 0.4, 0.66, 0.74, 0.78, 0.8, 0.81, 0.85, 0.85]

plt.close('all')

plt.xlabel("Months")
plt.ylabel("Conversion")

plt.plot(months, conversion)

ax = plt.subplot(1, 1, 1)
ax.set_xticks(months)
ax.set_xticklabels(month_names)
ax.set_yticks([0.10, 0.25, 0.5, 0.75])
ax.set_yticklabels(['10%','25%','50%','75%'])
plt.show()
````

````python

from matplotlib import pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]

# setting size of figure to 12 high and 4 wide
plt.figure(figsize=(12, 4)) 

#Subplot 1
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker = 'o')
plt.title('Page Visits Per Month')
plt.xlabel('Month')
plt.ylabel('Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

# Subplot 2
ax2 = plt.subplot(1,2,2)
plt.plot(x_values, key_limes_per_month,     color = 'green')
plt.plot(x_values, persian_limes_per_month, color = 'orange')
plt.plot(x_values, blood_limes_per_month,   color = 'red')
plt.title('Lime Sales By Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(['Key Limes', 'Persian Limes', 'Blood Limes'])
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)


plt.savefig('Sublime Limes Visits and Sales.png')
plt.show()
````
